```python
#Traduzir os nomes para o portugues
atletas.rename(columns ={'Player':'Jogador', 'Pos':'Posição',
'Age':'Idade', 'Tm':'Time', 'Games': 'jogos','Game_Started':'Jogos_iniciados',
'Minutes':'Minutos','Field_Goals':'Cesta','Field_Goals_Attempts':'Tentativa_de_cesta',
'3 points':'3_pontos', '2_Points':'2_pontos','Free_throws':'Arremessos_livres',
'Assistis':'Assistencia','Steals':'Bola_roubada','Blocks':'Bloqueio','Fauls':'Faltas','Points':'Pontos','Media de 1 3pontos':'media_trespto'},inplace=True)
```

```python
atletas.info()
```

```python
atletas['Minutos'] = atletas['Minutos'].astype(float)
atletas['Cesta']=atletas['Cesta'].astype(float)
atletas['Tentativa_de_cesta']=atletas['Tentativa_de_cesta'].astype(float)
atletas['3_pontos']=atletas['3_pontos'].astype(float)
atletas['2_pontos']=atletas['2_pontos'].astype(float)
atletas['Arremessos_livres']=atletas['Arremessos_livres'].astype(float)
atletas['Assistencia']=atletas['Assistencia'].astype(float)
atletas['Bola_roubada']=atletas['Bola_roubada'].astype(float)
atletas['Bloqueio']=atletas['Bloqueio'].astype(float)
atletas['Faltas']=atletas['Faltas'].astype(float)
atletas['Pontos']=atletas['Pontos'].astype(float)
```

```python
#analise por idade - tabela de frequências
numidade=atletas['Idade'].value_counts()
numidade
```

```python
def idade(idade):
  if idade <= 25:
    return ('Menor_que_25')
  if 26< idade < 30:
    return ('Entre_26_e_30')
  if 31< idade < 35:
    return ('Entre_31_e_35')
  if 36< idade < 40:
    return ('Entre_36_e_40')
```

```python
atletas.loc[:,'Intervalo_idade']=atletas['Idade'].apply(idade)
atletas.info()
```

```python

intervalo=atletas['Intervalo_idade'].value_counts()
#grafico de pizza
intervalo.plot.pie(title="Intervalo_idade",legend=True,autopct='%.0f%%',figsize=(7, 7))
```
